### <font color="Greenyellow">2. Creacion e INSERT de datos a nuestra base de datos en SQLite</font>
[Volver al indice](#item0)

```python
df = pd.read_csv(carpeta_datos + '/card_transdata.csv')
df
```

```python
# Abrimos la conexion a la base de datos
conexion_bbdd = sqlite3.connect(f'{carpeta_base_datos}/base_datos_tfm.db')
# Hacemos INSERT de nuestros datos a la base de datos
df.to_sql('mis_datos', conexion_bbdd, if_exists='replace', index=False)
# Cerramos la conexion
conexion_bbdd.close()
```
